```python
#   import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

titanic = sns.load_dataset("titanic")
titanic
```

```python
#   Understanding the distribution of Fares in Classes using scatter plot
plt.figure(figsize=(20,5))
sns.scatterplot(data=titanic, x="age", y="fare", hue="class")

#   From this graph, we can say that - 
#   most of the fare price lies in between 5-100 USD for Second and Third Class
#   most of the First Class fare price lies in between 100-200 USD
#   some of the First Class fare price lies in between 200-300 USD
#   two fares of First Class is over 500 USD
```

```python
#   Diving the dataset into class First, Second and Third

firstClass = titanic["class"] == 'First'
secondClass = titanic["class"] == 'Second'
thirdClass = titanic["class"] == 'Third'
```

```python
#   Histogram of First Class
plt.figure(figsize=(20,5))
sns.histplot(data=titanic[firstClass], x='age', y='fare', hue='who', bins=15)
```

```python
#   Histogram of Second Class
plt.figure(figsize=(20,5))
sns.histplot(data=titanic[secondClass], x='age', y='fare', hue='who', bins=15)
```

```python
#   Histogram of Third Class
plt.figure(figsize=(20,5))
sns.histplot(data=titanic[thirdClass], x='age', y='fare', hue='who', bins=15)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Titanic dataset from Seaborn
titanic_df = sns.load_dataset('titanic')

# Display the first few rows of the dataset
print(titanic_df.head())

# Plot histogram to visualize the distribution of ticket prices (fare)
plt.figure(figsize=(8, 6))
sns.histplot(titanic_df['fare'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Ticket Prices (Fare)')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
